### Crie uma função que efetue a leitura do arquivo e processe seus dados identificando o total de ocorrencias no período avaliado e, baseado nessa informação criar um arquivo de saída em formato JSON (statistics.json) com as seguintes as estatísticas (número de ocorrências e percentual do total) por fase de operação (coluna aeronave_fase_operacao), como o exemplo a seguir:

### Crie um arquivo de saida (formato CSV) com nome levels.csv contendo as seguintes informações:
* operation -> aeronave_operador_categoria
* type -> aeronave_tipo_veiculo
* manufacturer -> aeronave_fabricante
* engine_type aeronave_motor_tipo
* engines -> aeronave_motor_quantidade
* year_manufacturing -> aeronave_ano_fabricacao 
* seating -> aeronave_assentos
* fatalities -> total_fatalidades

### Considerando apenas acidentes cujo nível de dano da aeronave tenha sido LEVE ou NENHUM (coluna aeronave_nivel_dano) E que o número de fatalidades (total_fatalidades) tenha sido superior à 0 (zero)
OBS: 
1. Efetuar apenas uma leitura do arquivo de entrada
2. Efetuar todas as leituras e gravações de arquivos na pasta /tmp


````python
import csv
import json
import datetime

def read_csv(start_date, end_date):
    
    def save_csv(list):
        # csv writer (levels)
        with open('levels.csv',mode='w', newline='') as le:
            headers = ['operation', 'type', 'manufacturer', 'engine_type', 'engines', 'year_manufacturing', 'seating', 'fatalities']
            writer = csv.DictWriter(le, fieldnames=headers, delimiter=';', quoting=csv.QUOTE_ALL)
            writer.writeheader()
            writer.writerows(list)
    
    def create_statistic_list(stat_total, total):
        static = {}
        static_list = []
        for stat in stat_total.keys():
            static["fase_operacao"] = stat
            static["ocorrencias"] = stat_total.get(stat)
            perc = static["ocorrencias"] / total * 100
            static["percentual"] = "%.2f" % round(perc,2) + '%'
            static_list.append(static.copy())
        return static_list
    
    # variables
    stat_total = {}
    csv_dict = {}
    csv_list = []
    statistic_list = []
    total = 0
    filename = "anv.csv"
    
    # open input file for reading
    with open(filename, mode='r', encoding="utf8") as f:
        # csv reader
        reader = csv.DictReader(f,delimiter='~')
       
        # dates
        date_time_obj = datetime.datetime.strptime(start_date, '%d/%m/%Y')
        date_ini = date_time_obj.date() 

        date_time_obj = datetime.datetime.strptime(end_date, '%d/%m/%Y')
        date_fim = date_time_obj.date() 
        
        # iterate over file lines
        for line in reader:
            data = line['aeronave_dia_extracao']
            data_obj = datetime.datetime.strptime(data, '%Y-%m-%d')
            if (data_obj.date() >= date_ini) and (data_obj.date() <= date_fim):
                if (line['aeronave_nivel_dano'].lower() in ('leve','nenhum')):
                    if (line['total_fatalidades'] > '0'):                    
                        total += 1
                        # Qtty per operation phase
                        if stat_total.get(line['aeronave_fase_operacao']) is None:
                            stat_total[line['aeronave_fase_operacao']] = 1
                        else:
                            stat_total[line['aeronave_fase_operacao']] += 1

                        # add line into csv level file
                        csv_dict['operation'] = line['aeronave_operador_categoria']
                        csv_dict['type'] = line['aeronave_tipo_veiculo']
                        csv_dict['manufacturer'] = line['aeronave_fabricante']
                        csv_dict['engine_type'] = line['aeronave_motor_tipo']
                        csv_dict['engines'] = line['aeronave_motor_quantidade']
                        csv_dict['year_manufacturing'] = line['aeronave_ano_fabricacao'] 
                        csv_dict['seating'] = line['aeronave_assentos']
                        csv_dict['fatalities'] = line['total_fatalidades']
                        csv_list.append(csv_dict.copy())
        
        # save file levels.csv
        save_csv(csv_list)
        
        # read keys and values and add into the list
        statistic_list = create_statistic_list(stat_total, total)
                    
        # create statistics json file
        with open('statistics.json', 'w') as outfile:
            json.dump(statistic_list, outfile)
        
````

## Exercício 2

### Crie um programa que efetue a leitura do arquivo produzido no exercício 1 (levels.csv) e mostre na tela as seguintes informações estatísticas:
* Quantidade total de acidentes
* Quantidade total de acidentes agrupados por tipo de aeronave (type)
* Ano e fabricante da aeronave mais antiga (year_manufacturing, manufacturer)
* Quantidade de assentos e tipo de motor da aeronave mais nova (seating, engine_type)
* A quantidade de acidentes que ocorreram com aeronaves do tipo (type) HELICÓPTERO, cuja fabricação se deu após o ano 1997
* A quantidade de acidentes que ocorreram com aeronaves do tipo (type) AVIÃO fabricadas (manufacturer) pela CESSNA AIRCRAFT

OBS: 
1. Efetuar todas as leituras e gravações de arquivos na pasta /tmp

````python
import csv

def read_levels():
    # variables
    total = 0
    total_type = {}
    oldest = {"year_manufacturing": 9999, "manufacturer": ""}
    youngest = {"year": 0 , "seating": "", "engine_type": ""}
    helic_bef_97 = 0
    airplane_accident = 0
        
    filename = "levels.csv"
    
    # open input file for reading
    with open(filename, mode='r') as f:
        # csv reader
        reader = csv.DictReader(f,delimiter=';')
        
        # iterate over file lines
        for line in reader:
            total += 1
            
            # Qtty per type
            if total_type.get(line['type']) is None:
                total_type[line['type']] = 1
            else:
                total_type[line['type']] += 1
                
            if line['year_manufacturing'].upper() != "NULL":
                # Oldest manufacturing year
                if int(line['year_manufacturing']) <= oldest.get('year_manufacturing') and int(line['year_manufacturing']) > 0:
                    if int(line['year_manufacturing']) == oldest.get('year_manufacturing'):
                        oldest['manufacturer'] += ' | ' + line['manufacturer']
                    else:
                        oldest['manufacturer'] = line['manufacturer']
                    oldest['year_manufacturing'] = int(line['year_manufacturing'])

                # Youngest aircraft
                if int(line['year_manufacturing']) >= youngest.get('year') and line['seating'] != 'NULL':
                    if int(line['year_manufacturing']) == youngest.get('year'):
                        youngest['seating'] += '|' + line['seating']
                        youngest['engine_type'] += '|' + line['engine_type']
                    else:                    
                        youngest['seating'] = line['seating']
                        youngest['engine_type'] = line['engine_type']
                    youngest['year'] = int(line['year_manufacturing'])

                # Helicóptero with manufacturing year > 1997    
                if (int(line['year_manufacturing']) > 1997) and (line['type'].lower() == 'helicóptero'):
                    helic_bef_97 += 1 
            
            # Airplane
            if (line['type'].lower() == 'avião') and (line['manufacturer'].upper() == 'CESSNA AIRCRAFT'):  
                airplane_accident += 1
        
    print("ACIDENTES")
    print("=========")
    print("Total de Acidentes: " + str(total))
    print("--------------------------------")    
    print("Aeronave(s) mais antiga: ")
    print(">>> Ano: " + str(oldest['year_manufacturing']))
    print(">>> Fabricante: " + oldest['manufacturer'])
    print("--------------------------------")    
    print("Aeronave(s) mais nova: ")
    print(">>> Qtde de Assentos: " + youngest['seating'])
    print(">>> Tipo de Motor: " + youngest['engine_type'])
    print("--------------------------------")    
    print("Total de Helicópteros contruídos depois de 1997: " + str(helic_bef_97))
    print("--------------------------------")    
    print("Total de aviões da marca CESSNA AIRCRAFT: " + str(airplane_accident))
    print("--------------------------------")    
    print("Total de Acidentes por tipo:")
    for key in total_type.keys():
        print(">>> Tipo: " + key)
        print(">>> Qtde: " + str(total_type.get(key)))
        print("------------------")
````

### Crie uma função que efetue a leitura do arquivo (cota_parlamentar_sp.csv) e para cada legenda (sgpartido) grave um arquivo com o nome da legenda no formato JSON (ex: DEM.json, PSDB.json)

### Em cada arquivo deverá conter as seguintes informações conforme exemplo:

```json
[
  {
    "legenda": "DEM",
    "nome": "ROBSON TUMA",
    "total": 177565
  },
  {
    "legenda": "DEM",
    "nome": "JORGINHO MALULY",
    "total": 31335515
  },
  {
    "legenda": "SD",
    "nome": "PAULO PEREIRA DA SILVA", 
    "total": 155445892
  }
]
```

* legenda (partido)
* nome (nome do parlamentar)
* total (total de despesas do parlamentar)

OBS: 
1. Efetuar todas as leituras e gravações de arquivos na pasta /tmp

````python
import csv
import json

def split_parties():
    
    def save_party_file(name, deputies):
        partie_aux = []
        info = {"legenda": name, "nome": "", "total": 0}
        for depute in deputies.keys():
            info["nome"] = depute
            info["total"] += deputies.get(depute)
            partie_aux.append(info.copy())

        # create party json file
        with open(name + '.json', 'w') as outfile:
            json.dump(partie_aux, outfile)   
            
    # variables
    parties = {}
    filename = "cota_parlamentar_sp.csv"
    
    # open input file for reading
    with open(filename, mode='r', encoding="utf8") as f:
        # csv reader
        reader = csv.DictReader(f,delimiter=';')
        
        # iterate over file lines
        for line in reader:
            # Prepare party legend    
            if parties.get(line['sgpartido']) is None:
                parties[line['sgpartido']] = {}
                        
            # Prepare deputies name and sum values
            if parties.get(line['sgpartido']).get(line['txnomeparlamentar']) is None:
                parties.get(line['sgpartido'])[line['txnomeparlamentar']] = float(line['vlrdocumento'])
            else:
                parties.get(line['sgpartido'])[line['txnomeparlamentar']] += float(line['vlrdocumento'])
        
        for party in parties.keys():
            save_party_file(party, parties.get(party))
````
